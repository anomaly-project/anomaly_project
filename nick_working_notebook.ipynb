```python
def get_log_data():
    '''
    This function is to get the log dataset from a local csv file or from SQL Ace to our working notebook to be able to
    use the data and perform various tasks using the data
    ''' 
    
    if os.path.isfile('curriculum_logs.csv'):
        
        return pd.read_csv('curriculum_logs.csv')
    
    else:
       
        url = get_connection('curriculum_logs')
        
        query = '''
        SELECT *
        FROM logs as l
        JOIN cohorts as c
        ON c.id = l.cohort_id;
        '''

        df = pd.read_sql(query, url)
        
        df.to_csv('curriculum_logs.csv', index = False)

        return df   
```

```python
def get_ip():
    response = requests.get('https://api64.ipify.org?format=json').json()
    return response["ip"]


def get_location(ip_address):
    response = requests.get(f'https://ipapi.co/{ip_address}/json/').json()
    location_data = {
        "ip": ip_address,
        "city": response.get("city"),
        "region": response.get("region"),
        "country": response.get("country_name")
    }
    return location_data
```

```python
# Apply get_location function to each IP address in the 'ip' column
locations = ip_df['ip'].apply(lambda x: get_location(x))
```
