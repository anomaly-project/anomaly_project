### 3. Are there students who, when active, hardly access the curriculum? If so, what information do you have about these students?

```python
df = a.get_log_data()
df = p.clean_logs(df)
```

```python
#Count on path by cohort_id, merge dataframes, and create columns for active users
pages_by_user = pd.DataFrame(df.groupby(['cohort_id']).path.count().reset_index()).rename(columns={
'path':'pages_visited'})
df = df.merge(pages_by_user, how='left', on=['cohort_id']).set_index(df.index)
df['active'] = (df.index >= df['start_date']) & (df.index <= df['end_date'])

# Calculate the average pages visited per user
user_avg_pages = df.groupby('user_id')['pages_visited'].mean().reset_index()
```

```python
#k
multiplier =1.5

#calculate our q1 and q3
q1 = df.pages_visited.quantile(0.25)
q3 = df.pages_visited.quantile(0.75)
iqr = q3-q1

#calculate our fences
inner_lower_fence = q1 - (multiplier * iqr) 
inner_upper_fence = q3 + (multiplier * iqr) 

# Identify the users who visited significantly fewer pages than the average
low_activity_users = user_avg_pages[user_avg_pages['pages_visited'] < 9000]

# Merge with the original DataFrame to get more information about these users
low_activity_data = pd.merge(low_activity_users['user_id'], df, on='user_id', how='left')

#active and inactive laggers
df_without_duplicates = low_activity_data.drop_duplicates(subset=['user_id'])
active_laggers = df_without_duplicates[df_without_duplicates['active'] == True]
inactive_laggers = df_without_duplicates[df_without_duplicates['active'] == False]
```

**Takeaways**

 - We have about 103 active users who don't seem to visit our curriculum as much

### 4. Is there any suspicious activity, such as users/machines/etc accessing the curriculum who shouldn’t be? Does it appear that any web-scraping is happening? Are there any suspicious IP addresses?

```python
# Count the number of pages visited by each IP address
pages_by_ip = df.groupby('ip')['path'].count().sort_values(ascending=False)

#start with an inner fence calculation
multiplier = 3

#calculate our q1 and q3
q1 = pages_by_ip.quantile(0.25)
q3 = pages_by_ip.quantile(0.75)
iqr = q3-q1

# Fences
inner_lower_fence = q1 - (multiplier * iqr) 
inner_upper_fence = q3 + (multiplier * iqr)

#make col into data frame and rename 
pages_by_ip = pd.DataFrame(pages_by_ip)
pages_by_ip.rename(columns={'path': 'ip_pages_visited'}, inplace = True)

# merge data frame to original data frame
df = pd.merge(pages_by_ip, df, on = 'ip', how = 'left')

# drop dulplicated ip address and subset them as well 
ip_df = df.drop_duplicates(subset=['ip'])

# suspicious ip address
sus_ip = ip_df[ip_df['ip_pages_visited'] > 50]
```

**Takeaways**

 - Out of about 5000 unique IP addresses, we would flag about 1000 or 20% of them.
 
 - I would really focus on the top 4 as they seem high in the amount of times they access our curriculum
